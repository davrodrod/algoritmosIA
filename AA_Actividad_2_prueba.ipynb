
### Determinar el conjunto de modelización y el de validación. 

En primer lugar se elimina la variable objetivo 'price_range' del conjunto de variables, y se transforman los datos con la función 'StandardScaler' para que todos ellos tomen unos rangos similares.


```python
from sklearn.preprocessing import StandardScaler

#Se elimina la variabla price_range del X
X = X.drop(['price_range'], axis = 1)

#Normalización de los datos
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X))
```

Se utiliza la función "train_test_split" para dividir los datos entre conjuntos de entrenamiento y de test de manera aleatoria. Se utiliza un 80% de muestras para el entrenamiento y un 20% para el test.

Para la variable objetivo 'y' se codifica con oneHotEncoder de forma que, al haber 4 categorías, el resultado final tiene 4 columnas, una para cada categoría. Esto se utiliza para redes neuronales

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

ohe = OneHotEncoder()
y2 = y.values.reshape(-1,1)
y_ohe = ohe.fit_transform(y2).toarray()
#Se realiza la división de los datos para SVM
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2,random_state=RANDOM_STATE)

#Se realiza la división de los datos para redes neuronales
train_X_nn, test_X_nn, train_y_nn, test_y_nn = train_test_split(X, y_ohe, test_size=0.2,random_state=RANDOM_STATE)
```

### Calcular las métricas de evaluación de ajuste adecuadas:

Se realiza una funcion genérica para comparar el performance de todos los clasificadores. Esta función muestra:

* Matriz de confusión
* Métricas totales
* Métricas en cada una de las clases (grupos)
* Modelo utilizado
* Gráfica ROC por clases mostrando también el AUC

```python
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import label_binarize
from numpy import argmax

    
def performanceClassifier2(model, valX, valy):
    #model.fit(trainX, trainy)
    predicciones = model.predict(valX)
    y_pred_proba = model.predict_proba(valX)
    valy_binarizada = label_binarize(valy, model.classes_)
    if valy_binarizada.shape[1]==1:
        valy_binarizada = np.append(valy_binarizada*-1+1, valy_binarizada, axis=1)
    
    if valy_binarizada.shape[1]==1:
        valy_binarizada = np.append(valy_binarizada*-1+1, valy_binarizada, axis=1)
   
    # Matriz de confusión
    print("\nMatriz de confusión: ")
    print("==================== ")
    if (predicciones.ndim>1):
        predlist = list()
        for i in range(len(predicciones)):
            predlist.append(np.argmax(predicciones[i]))
        predicciones=pd.Series(predlist)
    #matrizConfusion = confusion_matrix(valy, predicciones)
    #print(matrizConfusion)
    print(pd.crosstab(valy, predicciones, rownames=["Real"], colnames=["Predicho"]))
    # Métricas
    print('\nModel Performance:')
    print('==================')
    print("\nAccuracy Total: ", metrics.accuracy_score(valy, predicciones))
    print("Precision Total: ", metrics.precision_score(valy, predicciones, average='weighted'))
    print("Recall Total: ", metrics.recall_score(valy, predicciones, average='weighted'))
    print("F1 score Total: ", metrics.f1_score(valy, predicciones, average='weighted'))
    
    classes = valy.unique()
    n_classes = classes.size
    
    # Métricas para cada clase
    for i in classes:
        print("\nAccuracy %s: %0.6f" % (i,metrics.accuracy_score(valy[valy==i], predicciones[valy==i])))
        print("Precision %s: %0.6f"% (i,metrics.precision_score(valy[valy==i], predicciones[valy==i], average='weighted')))
        print("Recall %s: %0.6f" % (i,metrics.recall_score(valy[valy==i], predicciones[valy==i], average='weighted')))
        print("F1 score %s: %0.6f" % (i,metrics.f1_score(valy[valy==i], predicciones[valy==i], average='weighted')))
    
    # Modelo utilizado
    print("\nModelo clasificador:")
    print("====================")
    print(model)
      
    # ROC y AUC
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(valy_binarizada[:, i], y_pred_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    for i in range(n_classes):
        plt.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %s' % (roc_auc[i], model.classes_[i]))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.legend(loc='lower right')
    plt.title("ROC Curve")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    
    print('\nGráficas: ROC:')
    print('===============================================')
    
    plt.show()
```

##### Clasificación con máquina de vector soporte

Para comparar resultados se utiliza primero el algoritmo de Máquina de Vector Soporte sin validación cruzada.

```python
from sklearn.svm import SVC

svm=SVC(random_state=RANDOM_STATE, probability=True)
svm.fit(train_X,train_y)
print("train accuracy:",svm.score(train_X, train_y))
print("test accuracy:",svm.score(test_X,test_y))

performanceClassifier2(svm, test_X, test_y)
```

Tomando por ejemplo el accuracy o exactitud como valor a observar, a priori el resultado con el algoritmo SVM sin validación cruzada es bastante aceptable, se aproxima al 90%.

Se observa una diferencia bastante alta entre los resultados en el conjunto de entrenamiento y el conjunto de test, parece que hay algo de overfitting. 

Ahora se repite con validación cruzada, utilizando 5 folds. Además se realiza un búsqueda de rejilla sobre un conjunto de parámetros.

```python
from sklearn.model_selection import GridSearchCV

kernel=["linear","rbf"]
decision_function_shape=["ovo","ovr"]

svm=SVC(random_state=RANDOM_STATE, probability=True, verbose=True)
grid_svm=GridSearchCV(estimator=svm,cv=5,param_grid=dict(kernel=kernel,decision_function_shape=decision_function_shape))
grid_svm.fit(X=train_X,y=train_y)
print("best score: ", grid_svm.best_score_)
print("best param: ", grid_svm.best_params_)

# Se selecciona el mejor clasificador
best_svm_model = grid_svm.best_estimator_

performanceClassifier2(best_svm_model, test_X, test_y)
```

En primer lugar se observa que el algoritmo es costoso computacionalmente. Si se dejan demasiados parámetros para explorar en la rejilla, va bastante lento. Se dejan sólo dos conjuntos de datos.

El resultado en accuracy es del 97% en el conjunto de datos de test.
